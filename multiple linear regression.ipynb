```python
x = np.array(data.iloc[0: ,1:])
y = np.array(data.iloc[0: ,0:1])
x.shape
```

```python
def loss_function(w,b,x,y):
    total_error = 0
    ##for i in range(len(x)):
        ##xd = x[i]
        ##yd = y[i]
    y_hat=[]
    for i in range(len(x)):
        y_hat.append(np.dot(w,x[i]) + b)
    y_hat=np.array(y_hat)
    total_error = np.sum((y_hat-y)**2)
        
        ##total_error += (yd - (np.dot(w,xd) + b)) ** 2
    total_cost = total_error/(len(x))
    return total_cost
```

```python
def gd(x,y,l,w_now,b_now):
    w_gradient = np.zeros(12)
    b_gradient = 0
    
    n = len(x)
    
    for i in range(len(x)): 
        xd = x[i]
        yd = y[i]
        
        w_gradient += (1/n)*xd*(yd - (np.dot(w_now,xd) + b_now))
        b_gradient += (1/n)*(yd-(np.dot(w_now,xd) + b_now))
    
    w = w_now - l*w_gradient
    b = b_now - l*b_gradient
    return w,b 
    
    
    
    
```

```python
w = np.zeros(12)
b = 0
epochs = 1000
l = 0.0000000001


for i in range(epochs):
    w,b = gd(x,y,l,w,b,)
    cost = loss_function(w,b,x,y)
    if i % 50 == 0:
        print(f"epochs: {i}")
        print(cost)
        print(w)

print(w,b)

        
```
